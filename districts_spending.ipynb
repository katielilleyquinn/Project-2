```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "districts-spending.csv"
districts_data_df = pd.read_csv(csv_file)
districts_data_df
```

```python
twentysixteen_df = districts_data_df.loc[districts_data_df["YRDATA"] == 2016, :]
twentysixteen_df
```

```python
reduced_df = twentysixteen_df.drop(columns=['TFEDREV', 'TSTREV', 'TLOCREV', 'TCURINST', 'TCURSSVC', 'TCURONON', 'TCAPOUT'])
reduced_df
```

```python
PL = reduced_df['TOTALREV'] - reduced_df['TOTALEXP']
reduced_df['PL'] = PL
reduced_df
```

```python
final_df = reduced_df.drop(columns=['Profit/Loss', 'PL'])
final_df
```

```python
df_group = final_df.groupby(["STATE"])
avg_df = df_group.mean()
avg_df
```

```python
avg_df.to_csv('StatesPL.csv', index = True, header=True)
```
